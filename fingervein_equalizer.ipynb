```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("fingervein.bmp",cv2.IMREAD_GRAYSCALE) #reading fingervein as grayscale
```

```python
imgplot = plt.imshow(img , cmap="gray")
plt.title("Finger vein image")
plt.show()
```

```python
plt.hist(img.ravel(),bins=256)
plt.title("Histogram of Original image")
plt.show()
```

```python
histeql = cv2.equalizeHist(img)
```

```python
plt.imshow(histeql,cmap='gray')
plt.title("Equalized Image")
plt.show()
```

```python
plt.hist(histeql.ravel(),bins=256)
plt.title("Histogram of Equalized Image")
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=4.0,tileGridSize=(16,16))
```

```python
clahe_img = clahe.apply(img)
```

```python
plt.imshow(clahe_img,cmap='gray')
plt.title("Histogram Equalized Image")
plt.show()
```

```python
plt.hist(clahe_img.ravel(),bins=256)
plt.title("Histogram of CLAHE Image")
plt.show()
```
